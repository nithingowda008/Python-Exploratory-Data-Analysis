DATA CLEANING

```python
# STEP 1: FILTERING DATA 
# Extraxt only races of Country = USA, Distance = 50km/50mi and year = 2020
```

```python
#show Event distance/length = 50km or 50mi      
df[df['Event distance/length']== '50km']
df[df['Event distance/length']== '50mi']
```

```python
# Combine 50km and 50mi using isin()

df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
# Extract only races for year=2020
```

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020)]
```

```python
# Extract races where Athlete country = USA

df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
                                                                                                        
```

```python
# Combining all filters

df2=df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020) & 
   (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

```python
# Remove USA in Event name Column

df2['Event name'].str.split("(").str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split("(").str.get(0)
```

```python
df2.head()
```

```python
# Clean up errors in Athlete's age column by adding new column

df2['atlete_age'] = 2020 -df2['Athlete year of birth']
```

```python
# remove 'h' from athletes performance column

df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
# drop columns : Athlete club, Athlete ciuntry, athlete Dob, Athlete age category

df2 = df2.drop(columns=['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2.head()
```

```python
df2.shape
```

```python
# Clean up null values

df2.isna().sum()
```

```python
# 233 nulls in atlete_age

df2[df2['atlete_age'].isna() == 1]

# showing rows with null values
```

```python
# Dropping null values

df2 = df2.dropna()
```

```python
df2.shape # 233 rows removed
```

```python
# Check for duplicates

df2[df2.duplicated()== True]

# NO Duplicates
```

```python
# Restting index

df2.reset_index(drop = True)
```

```python
# Change datatypes

df2.dtypes
```

```python
df2['atlete_age'] = df2['atlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df3=df2.rename(columns={
                        'Year of event': 'year',
                        'Event dates': 'race_year',
                        'Event name': 'race_name',
                        'Event distance/length': 'race_length',
                        'Event number of finishers': 'race_number_of_finishers',
                        'Athlete performance': 'athlete_performance',
                        'Athlete gender': 'gender',
                        'Athlete average speed': 'athlete_avg_speed',
                        'Athlete ID': 'athlete_ID',
                        'atlete_age': 'atlete_age'
                        })
```

```python
df3.head()
```

DATA ANALYSIS AND VISUALISATION

```python
# 1. Number of athletes in 50km and 50mi races

plt.hist(df2['race_length'], bins=3, edgecolor = 'black')
plt.xlabel('Race Length')
plt.ylabel('COUNT')
plt.title('Histogram of race length')
plt.show()
```

```python
import seaborn as sns
```

```python
#2. Average speed of Athletes
sns.distplot(df2[df2['race_length']== '50mi']['athlete_avg_speed'])
plt.xlabel('AVG_SPEED')
plt.ylabel('COUNT')
plt.title('Distribution plot')
```

```python
#3. Violin plot to analyse AVG_SPEED of males v/s females over 50km and 50mi races

sns.violinplot(data=df2, x='race_length', y= 'athlete_avg_speed', hue ='gender', split = True, inner='quart')
plt.xlabel('race_length')
plt.ylabel('athlete_avg_speed')
plt.title('violin plot for avg_speed of males v/s females')

# Males are faater in 50km races while they are in similar brackets in 50mi 
# As the race lenghth increases females tend to have higher avg speed.
```

```python
# 4. Finding linear regression line using lmplot

sns.lmplot(data=df3, x='atlete_age', y='athlete_avg_speed', hue = 'gender')
plt.xlabel('atlete_age')
plt.ylabel('athlete_avg_speed')
plt.title('lmplot for age v/s avg_speed')
```

```python
sns.regplot(data=df3, x='atlete_age', y='athlete_avg_speed')
```

```python
#5. Difference in speed for the 50km/50mi race for males and females

df3.groupby(['race_length','gender'])['athlete_avg_speed'].mean()
```

```python
a= df3.groupby(['race_length','gender'])['athlete_avg_speed'].mean().reset_index()
```

```python
sns.barplot(data=a, x='race_length', y='athlete_avg_speed', hue='gender')
plt.xlabel('race_length')
plt.ylabel('athlete_avg_speed')
plt.title('Difference in speed for the 50km/50mi race for males and females')
plt.show()
```

```python
#6. What age groups are the best in 50mi races (min 10 races)

best=df3.query('race_length == "50mi"').groupby('atlete_age')['athlete_avg_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>9')
best.head()
```

```python
# What age groups are the Worst in 50mi races (min 10 races)

worst=df3.query('race_length == "50mi"').groupby('atlete_age')['athlete_avg_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>9')
worst.head()
```

```python
#7. Seasons for the race >> are runners slower in Summer than in winter ?

# spring = 3-5
# summer = 6-8
# autumn = 9-11
# winter = 12-2
```

```python
# using split in race_
df3.dtypes
df3['race_month']=df3['race_year'].str.split('.').str.get(1).astype(int)
```

```python
df3.head()
```

```python
# Using lambda function to assign seasons 

df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Autumn' if x > 8 else 'Summer' if x > 5 else 'Spring' if x>2 else 'Winter')
```

```python
df3.head()
```

```python
df3.groupby('race_season')['athlete_avg_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```

```python
Seasons=df3.groupby('race_season')['athlete_avg_speed'].agg(['mean','count']).sort_values('mean', ascending = False).reset_index()
```

```python
sns.lineplot(data=Seasons, x = 'race_season', y='mean')
```

```python
# Therfore, from above line chart summer season races have lowest avg_speeds among athletes.
```
